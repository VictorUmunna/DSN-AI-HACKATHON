## EXPLORATORY DATA ANALYSIS

```python
# distribution of the train year of birth column
fig = plt.figure(figsize=(20,10))
sns.histplot(data = df_train, x = 'Year_of_Birth', kde = True)
plt.title("Distribution of `the year of birth")
```

```python
df_train['Year_of_Birth'].unique()
```

1900 is an outlier so we remove it

```python
df_train[df_train['Year_of_Birth']== 1900]
```

```python
df_train = df_train.loc[df_train['Year_of_Birth'] != 1900]
```

```python
df_train[df_train['Year_of_Birth']== 1900]
```

```python
# distribution of the train year of birth column
fig = plt.figure(figsize=(20,10))
sns.histplot(data = df_train, x = 'Year_of_Birth', kde = True)
plt.title("Distribution of the year of birth")
```

Now, we have a normal distribution in the year of birth column

```python
# distribution of the test year of birth column
fig = plt.figure(figsize=(20,10))
sns.histplot(data = df_test, x = 'Year_of_Birth', kde = True)
plt.title("Distribution of the year of birth")
```

```python
df_test['Year_of_Birth'].unique()
```

Years 1901 and 1894 are the outliers

```python
df_test[(df_test['Year_of_Birth']== 1901) | (df_test['Year_of_Birth'] == 1894)]
```

```python
df_test = df_test.loc[df_test['Year_of_Birth'] != 1901] 
df_test = df_test.loc[df_test['Year_of_Birth'] != 1894]
```

```python
# distribution of the test year of birth column (test data)
fig = plt.figure(figsize=(20,10))
sns.histplot(data = df_test, x = 'Year_of_Birth', kde = True)
plt.title("Distribution of the year of birth")
```
