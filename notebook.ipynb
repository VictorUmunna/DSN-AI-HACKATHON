## EXPLORATORY DATA ANALYSIS

```python
# distribution of the train year of birth column
fig = plt.figure(figsize=(20,10))
sns.histplot(data = train_data, x = 'Year_of_Birth', kde = True)
plt.title("Distribution of `the year of birth of the training data")
```

```python
train_data['Year_of_Birth'].unique()
```

1900 is an outlier so we remove it

```python
train_data[train_data['Year_of_Birth']== 1900]
```

```python
train_data = train_data.loc[train_data['Year_of_Birth'] != 1900]
```

```python
# distribution of the train year of birth column
fig = plt.figure(figsize=(20,10))
sns.histplot(data = train_data, x = 'Year_of_Birth', kde = True)
plt.title("Distribution of the year of birth of the training dataset")
```

Now, we have a normal distribution in the year of birth column

```python
# distribution of the test year of birth column
fig = plt.figure(figsize=(20,10))
sns.histplot(data = test_data, x = 'Year_of_Birth', kde = True)
plt.title("Distribution of the year of birth of thee test dataset")
```

```python
test_data['Year_of_Birth'].unique()
```

Years 1901 and 1894 are the outliers

```python
test_data[(test_data['Year_of_Birth']== 1901) | (test_data['Year_of_Birth'] == 1894)]
```

```python
test_data = test_data.loc[test_data['Year_of_Birth'] != 1901] 
test_data = test_data.loc[test_data['Year_of_Birth'] != 1894]
```

```python
# distribution of the test year of birth column (test data)
fig = plt.figure(figsize=(20,10))
sns.histplot(data = test_data, x = 'Year_of_Birth', kde = True)
plt.title("Distribution of the year of birth of the test dataset")
```
